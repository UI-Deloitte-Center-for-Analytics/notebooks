Find the unique set of strings in the `michelin_award` column.

```python
df['michelin_award'].unique()
```

Create a function to convert the number of star strings to integers.

```python
def stars_text_to_number(stars_text):
    if stars_text == "ONE_STAR":
        return 1
    elif stars_text == "TWO_STARS":
        return 2
    else:
        return 3
```

Use the new function to convert the stars into numbers.

```python
df['michelin_award'].apply(stars_text_to_number)
```

Alternatively, use `np.where` to convert the strings in a one-liner.

```python
df['stars'] = np.where(df['michelin_award'] == 'ONE_STAR', 1, np.where(df['michelin_award'] == 'TWO_STARS', 2, 3))
```

### Number of restaurants, average number of stars, and total number of stars by country

```python
df_by_country = df.groupby('country', as_index=False).aggregate({
    'stars': ['count', 'sum', 'mean']
})

df_by_country.head(5)
```
